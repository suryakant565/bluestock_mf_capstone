```python
import pandas as pd
import sqlite3

conn = sqlite3.connect("../data/db/bluestock_mf.db")

for table in [
    "dim_fund",
    "fact_nav",
    "fact_transactions",
    "fact_performance",
    "fact_aum"
]:
    cnt = pd.read_sql(
        f"SELECT COUNT(*) as rows FROM {table}",
        conn
    )
    print(table)
    print(cnt)
```
